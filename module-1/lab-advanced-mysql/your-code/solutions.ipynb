```python
df = pd.read_sql_query("""
SELECT t.title_id, ta.au_id, SUM(t.price * s.qty * t.royalty / 100 * ta.royaltyper / 100) AS ROYALTIES
	FROM titles AS t
    RIGHT JOIN titleauthor AS ta ON t.title_id=ta.title_id
    LEFT JOIN sales AS s ON ta.title_id= s.title_id
    GROUP BY ta.au_id, t.title_id
""", engine)
df
```

```python
df2 = pd.read_sql_query("""
SELECT au_id, title, tl.advance, royal
FROM (SELECT t.title_id, ta.au_id, SUM(t.price * s.qty * t.royalty / 100 * ta.royaltyper / 100) AS royal
		FROM titles AS t
		RIGHT JOIN titleauthor AS ta ON t.title_id=ta.title_id
		LEFT JOIN sales AS s ON ta.title_id= s.title_id
		GROUP BY ta.au_id, t.title_id) as q
        RIGHT JOIN titles as tl ON tl.title_id = q.title_id
        order by royal desc
        LIMIT 3
""", engine)
df2
```

```python
df_temporal = pd.read_sql_query("""
    CREATE TEMPORARY TABLE  temporal3(
        SELECT t.title_id, ta.au_id, SUM(t.price * s.qty * t.royalty / 100 * ta.royaltyper / 100) AS NEW
            FROM titles AS t
            RIGHT JOIN titleauthor AS ta ON t.title_id=ta.title_id
            LEFT JOIN sales AS s ON ta.title_id= s.title_id
            GROUP BY ta.au_id, t.title_id)
""", engine)
```

```python
df3 = pd.read_sql_query("""
    SELECT tm.au_id, tm.title_id, tl.advance, tm.NEW	
	FROM temporal3 as tm
    RIGHT JOIN titles as tl ON tl.title_id = tm.title_id
	order by NEW desc
	LIMIT 3
""", engine)
df3
```
