```python
df['TotalSize'] = df[["GrLivArea","LotArea", "GarageArea"]].sum(axis=1)
```

```python
size = sns.jointplot(data=df, x='SalePrice', y='TotalSize', kind='reg', color='g')
plt.show()
```

```python
#Compara como influye el nÂº de cocinas, baÃ±os y habitaciones por separado y juntas
```

```python
df["TotalBath"] = df[["BedroomAbvGr", "KitchenAbvGr"]].sum(axis=1)
```

```python
data_baths = df.groupby(["TotalBath"]).agg({"SalePrice":"mean"})
data_baths.plot()
data_bedroom = df.groupby(["BedroomAbvGr"]).agg({"SalePrice":"mean"})
data_bedroom.plot()
data_kitchen = df.groupby(["KitchenAbvGr"]).agg({"SalePrice":"mean"})
data_kitchen.plot()
```

```python
data_Rooms = df.groupby(["TotalRooms"]).agg({"SalePrice":"mean"})
data_Rooms.plot(figsize=(20,10))
```

```python
#Separo los aÃ±os desde que se construyÃ³ y se remodelÃ³ y hago 2 grÃ¡ficas separadas. 
```

```python
df["YearTotalBuilt"]= 2020 - df["YearBuilt"]
```

```python
df["YearTotalRemod"]= 2020 - df["YearRemodAdd"]
```

```python
data_Built = df.groupby(["YearTotalBuilt"]).agg({"SalePrice":"mean"})
data_Remod= df.groupby(["YearTotalRemod"]).agg({"SalePrice":"mean"})
data_Built.plot(figsize=(20,10))
data_Remod.plot(figsize=(20,10))
```

```python
#Ver que vecindario tiene un precio mÃ¡s alto. 
```

```python
data_Neighborhood = df.groupby(["Neighborhood"]).agg({"SalePrice":"mean"})
data_Neighborhood = data_Neighborhood.sort_values(by=['SalePrice'], ascending=True)
data_Neighborhood.plot(kind="barh")
```

## Bonus Challenge 2 - Exploring Data with Correlation Heatmap

Now you have explored data visualizations with certain fields based on common sense. In the dataset there are many other fields that you are not sure whether they are important factors for the sale price. What is the best way to explore those fields without investigating them individually?

Making scatter matrix is not an option here because there are too many fields which makes it extremely time consuming to create scatter matrix. One option you have is to create a heatmap. Heatmaps are much less expensive to create than scatter matrixes. You can use heatmaps to visualize the pairwise correlations between each two variables.

Here is a [reference](https://seaborn.pydata.org/examples/many_pairwise_correlations.html) you can use to learn how to creat the pairwise correlation heatmap. Your heatmap should look like below:

![Corr Heatmap](heatmap.png)

```python
corr = df.corr()
mask = np.triu(np.ones_like(corr, dtype=np.bool))
f, ax = plt.subplots(figsize=(11, 9))
cmap = sns.diverging_palette(220, 10, as_cmap=True)
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

In your heatmap, you can easily identify the highly correlated (either positively or negatively) variables by looking for the grids with darker colors.
